# What turned into Renewable Area?

This notebook generates the plot, expressing what area that turned into renewable energy area in the world.

```python
%load_ext autoreload
%autoreload 2
```

```python
from src.DataBaseManager import DBMS

dbms = DBMS()
changes = dbms.read("WHAT_TURNED_INTO_RENEWABLES", {})
```

```python
changes.groupby(
    ["country", "lulc_category_from", "lulc_category_to"]
).sum().sort_values("area_converted_to_renewables", ascending=False)
```

```python
# Group by 'country', 'lulc_category_to', and 'lulc_category_from' and sum the areas
grouped = changes.groupby(["country", "lulc_category_to", "lulc_category_from"]).sum()

sorted_grouped = grouped.sort_values(
    by=["country", "lulc_category_to", "area_converted_to_renewables"],
    ascending=[True, True, False],
)

top_3_each = sorted_grouped.groupby(["country", "lulc_category_to"]).head(3)
```
